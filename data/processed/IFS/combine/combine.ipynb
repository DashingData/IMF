```python
import pandas as pd

# Step 1: List of specific file paths
file_paths = [
    "/home/abhishek/Desktop/IMF/data/processed/IFS/Balance_of_payment/IFS_Balance_Of_Paymnet.csv",
    "/home/abhishek/Desktop/IMF/data/processed/IFS/Exchange_rates/IFS_Exchange_rates.csv",
    "/home/abhishek/Desktop/IMF/data/processed/IFS/internation_investment_Position/IFS_International_investment_position.csv",
    "/home/abhishek/Desktop/IMF/data/processed/IFS/International_liquidity/IFS_International_Liquidity.csv",
    "/home/abhishek/Desktop/IMF/data/processed/IFS/soverign_data/IFS_Sovereign_data.csv"
]

# Step 2: Read each file into a list of DataFrames
dfs = [pd.read_csv(path) for path in file_paths]

# Step 3: Combine all into one DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

# Step 4: Save the combined file
save_path = "/home/abhishek/Desktop/IMF/data/processed/IFS/combine/IFS_combined.csv"
combined_df.to_csv(save_path, index=False)

# Confirmation
print("✅ IFS data combined and saved to:")
print(save_path)
```
