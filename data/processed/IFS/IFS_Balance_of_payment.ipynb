```python
import numpy as np
import pandas as pd
import glob
```

```python
# Arjusting the file format 
# Arjusting the file /home/abhishek/Desktop/IMF/data/raw/ifs/Balance_of_Payments.csv


df = pd.read_csv("/home/abhishek/Desktop/IMF/data/raw/ifs/Balance_of_Payments.csv")
```

```python
df.head()
```

```python
# Step 1: List of columns to drop
columns_to_drop = ["DATASET", "SERIES_CODE", "OBS_MEASURE", "2024"]

# Step 2: Drop them from the DataFrame
df = df.drop(columns=columns_to_drop)


```

```python
df.head()
```

```python
# Step 2: Identify year columns (2010 to 2023)
year_columns = [str(year) for year in range(2010, 2024)]

# Step 3: Get ID columns (all columns except years)
id_columns = [col for col in df.columns if col not in year_columns]

# Step 4: Melt the dataframe into long format
df_long = df.melt(
    id_vars=id_columns,
    value_vars=year_columns,
    var_name="year",
    value_name="value"
)

# Step 5: Convert 'year' to int
df_long["year"] = df_long["year"].astype(int)
```

```python
df_long.head()
```

```python
df_long.to_csv("/home/abhishek/Desktop/IMF/data/processed/IFS/IFS_Balance_Of_Paymnet.csv", index=False)
```
