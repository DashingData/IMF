```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv("/home/abhishek/Desktop/IMF/data/raw/weo/weo_raw_dataset.csv")
```

```python
df.head()
```

```python
# Step 1: List of columns to drop
columns_to_drop = ["DATASET", "SERIES_CODE", "OBS_MEASURE"]

# Step 2: Drop them from the DataFrame
df = df.drop(columns=columns_to_drop)
```

```python
df.head()
```

```python
# Step 2: Automatically detect year columns (e.g., 2005â€“2024)
year_columns = [col for col in df.columns if col.isdigit() and len(col) == 4]

# Step 3: Get ID columns (non-year columns)
id_columns = [col for col in df.columns if col not in year_columns]

# Step 4: Melt to long format
df_long = df.melt(
    id_vars=id_columns,
    value_vars=year_columns,
    var_name="year",
    value_name="value"
)

# Step 5: Convert year to integer
df_long["year"] = df_long["year"].astype(int)
```
