```python



# Step 1: List of columns to drop
columns_to_drop = ["BOP_ACCOUNTING_ENTRY",
        "UNIT",
        "FREQUENCY",
        "SCALE",
        "TYPE_OF_TRANSFORMATION",
        "Year of Default",
        "Default Type"]

# Step 2: Drop them from the DataFrame
df = df.drop(columns=columns_to_drop)
```

```python
# Step 1: Combine both indicator columns
df['indicator'] = df['indicator'].fillna(df['INDICATOR'])

# Optional: Drop the old column
df = df.drop(columns=['INDICATOR'], errors='ignore')
```

```python
df.head(0)
```

```python
# Normalize column names to lowercase
df.columns = df.columns.str.lower()

# Now apply the filter
df = df[df['country'].notnull() & df['year'].notnull() & df['indicator'].notnull() & df['value'].notnull()]
```

```python
df.head()
```

```python
df_wide = df.pivot_table(
    index=["country", "year"],
    columns="indicator",
    values="value",
    aggfunc="first"
).reset_index()
```

```python
df.head()
```

```python
labels = df[["country", "year", "manual sovereign default label"]].dropna().drop_duplicates()

df_final = df_wide.merge(labels, on=["country", "year"], how="left")
df_final["defaulted"] = df_final["manual sovereign default label"].fillna(0).astype(int)
```

```python
df.head()
```

```python
# Step 5: Drop high-NaN columns
missing_ratio = df_final.isnull().mean()
cols_to_keep = missing_ratio[missing_ratio <= 0.5].index
df_final = df_final[cols_to_keep]

# Step 6: Fill remaining NaNs with median
df_final = df_final.fillna(df_final.median(numeric_only=True))
```
